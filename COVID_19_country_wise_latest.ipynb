```python
Active_mean = df['Active'].mean()
df['Active'].map(lambda p: p - Active_mean)
```

```python
#Metode Apply
Active_mean = df['Active'].mean()

def remean_points(row):
    row['Active'] = row['Active'] - Active_mean
    return row

df.apply(remean_points, axis='columns').head()
```
