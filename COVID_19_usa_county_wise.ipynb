```python
Active_mean = df['Deaths'].mean()
df['Deaths'].map(lambda p: p - Active_mean)
```

```python
#Metode Apply
Active_mean = df['Deaths'].mean()

def remean_points(row):
    row['Deaths'] = row['Deaths'] - Active_mean
    return row

df.apply(remean_points, axis='columns').head()
```
